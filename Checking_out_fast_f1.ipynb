```python
import fastf1 as ff1
from fastf1 import plotting
from matplotlib import pyplot as plt

plotting.setup_mpl()

ff1.Cache.enable_cache('cache')  # optional but recommended
ff1.Cache.clear_cache('cache',deep=True)
Q = ff1.get_session(2021, 'Monaco', 'Q')
laps = Q.load_laps(with_telemetry=True)
```

```python
nor_fastest = laps.pick_driver('NOR').pick_fastest().telemetry
ric_fastest = laps.pick_driver('RIC').pick_fastest().telemetry


fig, ax = plt.subplots(5, 1,sharex=True,figsize=(15,10))
ax[0].plot(nor_fastest['Distance'], nor_fastest['Speed'], color='red')
ax[0].plot(ric_fastest['Distance'], ric_fastest['Speed'], color='cyan')
ax[0].set_title("NOR vs RIC")
ax[0].set_ylabel("Speed")


ax[1].plot(nor_fastest['Distance'], nor_fastest['Brake'], color='red')
ax[1].plot(ric_fastest['Distance'], ric_fastest['Brake'], color='cyan')
ax[1].set_ylabel("Brake")



ax[2].plot(nor_fastest['Distance'], nor_fastest['Throttle'], color='red')
ax[2].plot(ric_fastest['Distance'], ric_fastest['Throttle'], color='cyan')
ax[2].set_ylabel("Throttle")


ax[3].plot(nor_fastest['Distance'], nor_fastest['RPM'], color='red')
ax[3].plot(ric_fastest['Distance'], ric_fastest['RPM'], color='cyan')
ax[3].set_ylabel("RPM")


ax[4].plot(nor_fastest['Distance'], nor_fastest['nGear'], color='red')
ax[4].plot(ric_fastest['Distance'], ric_fastest['nGear'], color='cyan')
ax[4].set_ylabel("RPM")


plt.show()
```
